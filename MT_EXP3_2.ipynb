```python
#PRAC 10
# C) Using Python OpenCV library , create a histogram of an image 

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('pics.png' , 1 ) 
img
img.ravel()
img.shape
img.ravel().shape
img.ravel().size
 
plt.hist(img.ravel(), bins=256, range = [0,255])
plt.show()

colors = ('b', 'g', 'r')
 
img_ravel = [img[:, :, 0].ravel(), img[:, :, 1].ravel(), img[:,:, 2].ravel()] 
plt.hist(img_ravel, color=colors, label=colors)
plt.legend()
plt.show()

colors = ('b', 'g', 'r')
img_ravel = [img[:, :, 0].ravel(), img[:, :, 1].ravel(), img[:,:, 2].ravel()] 
plt.hist(img_ravel, color=colors, label=colors, bins=256, range=[0,256])
plt.legend()
plt.show()

colors = ('b', 'g', 'r')      
plt.figure(figsize=(16, 9))
for i, color in enumerate(colors):
    histogram = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histogram, color=color)
plt.show()
 
histogram
```

```python
#PRAC 10
# D) Using Python OpenCV library , Transpose an Image

import cv2 
from google.colab.patches import cv2_imshow 

src = cv2.imread('pics.png' , 1 ) 
image = cv2.transpose(src)
cv2_imshow(image)
```

```python
#PRAC 10
# E) Using Python OpenCV library , split an image
import cv2 
from google.colab.patches import cv2_imshow

image = cv2.imread('pics.png' , 1 ) 
 
# Using cv2.split() to split channels of coloured image 
b,g,r = cv2.split(image)
  
# Displaying Blue channel image
# Blue colour is highlighted the most
cv2_imshow(b)
  
# Displaying Green channel image
# Green colour is highlighted the most
cv2_imshow( g)
  
# Displaying Red channel image
# Red colour is highlighted the most
cv2_imshow(r)
```
